```python
# I admit I received helped from outside sources. Lots of stackoverflow links...
# I loop through all the ratings but whenever I find a rating done by user 15 or 200, I grab the movie id and search in items
# I grab that and subtract the mean of the user when adding to the user_profiles. I also update the count. (Bottom 2 rows)
for i in range(len(ratings)):
    if (ratings.loc[i]['user id'] == 15):
        movie_idx = ratings.loc[i]['movie id'] - 1
        movie = item.iloc[movie_idx][5:24]
        for j in range(len(movie)):
            if movie[j] == 1:
                user_profiles.loc[0][j + 1] += (ratings.loc[i]['rating'] - user_15_mean)
                user_profiles.loc[2][j + 1] += 1
    elif (ratings.loc[i]['user id'] == 200):
        movie_idx = ratings.loc[i]['movie id'] - 1
        movie = item.iloc[movie_idx][5:24]
        for j in range(len(movie)):
            if movie[j] == 1:
                user_profiles.loc[1][j + 1] += (ratings.loc[i]['rating'] - user_200_mean)
                user_profiles.loc[3][j + 1] += 1

user_profiles
```

```python
# Calculate the mean of every category
for i in range(1, 20):
    if (user_profiles.loc[2][i] != 0):
        user_profiles.loc[0][i] /= user_profiles.loc[2][i]

for i in range(1, 20):
    if (user_profiles.loc[3][i] != 0):
        user_profiles.loc[1][i] /= user_profiles.loc[3][i]
            
user_profiles = user_profiles.drop([2, 3])
user_profiles
```

```python
# Finally, I calculate the cosine similarity and distance between the users. 
# Seems like they have very different taste
user15 = user_profiles.iloc[:1,1:]
user200 = user_profiles.iloc[1:2,1:]
print(metrics.pairwise.cosine_similarity(user15,user200))
print(metrics.pairwise.cosine_distances(user15,user200))
```
